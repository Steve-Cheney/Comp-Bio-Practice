```
# Q1d
# len = 8
# 13 sites
con = "TGTAAACA" #consensus seq
a = "TGTAAACA"
b = "TGTAAACA"
c = "GGTAAACA"
d = "GCAAAACA"
e = "TGTAAACA"
f = "TGTAAACA"
g = "GGGAAACT"
h = "GCAAAACA"
i = "TATAAACA"
j = "GGTAAACA"
k = "TGAAAATA"
l = "TGTAAACA"
m = "TGTACACA"
sites = [a,b,c,d,e,f,g,h,i,j,k,l,m]
def proba(pos):
  count = 0
  for q in range(13):
    if sites[q][pos] == "A":
      count += 1
  return "{:.2f}".format(count / 13)

def probc(pos):
  count = 0
  for q in range(13):
    if sites[q][pos] == "C":
      count += 1
  return "{:.2f}".format(count / 13)

def probt(pos):
  count = 0
  for q in range(13):
    if sites[q][pos] == "T":
      count += 1
  return "{:.2f}".format(count / 13)

def probg(pos):
  count = 0
  for q in range(13):
    if sites[q][pos] == "G":
      count += 1
  return "{:.2f}".format(count / 13)

def prob_all_a(num):
  out = ""
  for z in range(num):
    out += str(proba(z)) + "|"
  return out

def prob_all_c(num):
  out = ""
  for z in range(num):
    out += str(probc(z)) + "|"
  return out

def prob_all_t(num):
  out = ""
  for z in range(num):
    out += str(probt(z)) + "|"
  return out

def prob_all_g(num):
  out = ""
  for z in range(num):
    out += str(probg(z)) + "|"
  return out

print("---"+("-----"*8))
print("|A|"+str(prob_all_a(8)))
print("|-|"+"-----"*8)
print("|C|"+str(prob_all_c(8)))
print("|-|"+"-----"*8)
print("|T|"+str(prob_all_t(8)))
print("|-|"+"-----"*8)
print("|G|"+str(prob_all_g(8)))
print("---"+("-----"*8))

print("\nProbability of",con,"is:","{:.2f}".format(0.62*0.77*0.69*1*.92*1*0.92*0.92))

print("\nProbability of GGTAAACA is:","{:.2f}".format(0.38*0.77*0.69*1*0.92*1*0.92*0.92))
```

```
################################################
# HW4 Solution - Generate a random sequence of nucleotides       #
################################################


####### Possible Solution ##########

nucsV = ["a","c","g","t"];

#Nuc prob (A,C,G,T)
def createSeqGc(seqLen,nucProb):
	x = np.random.choice(nucsV, seqLen, p=nucProb);
	#str(x);
	return "".join(x);

y = createSeqGc(10,[0.4,0.1,0.1,0.4])
print(y)
```

```
################################################
# Question 3a - Generate 1000 promoter sequences, each 1000 nucleotides long    #
################################################
promote_len = 1000
nuc_len = 1000

promote_seqs_strep = []
promote_seqs_sacc = []

strep_p = [0.14,0.36,0.36,0.14]
sacc_p = [0.31,0.19,0.19,0.31]

for i in range(promote_len):
  promote_seqs_strep.append(createSeqGc(nuc_len,strep_p))
  promote_seqs_sacc.append(createSeqGc(nuc_len,sacc_p))
```

```
################################################
# Question 3b - Count the number of times the FOXO1 consensus is found in the promoters  #
################################################

foxo1Consensus = "gtaaac";
foxolen = len(foxo1Consensus)

motif_count_strep = 0
motif_count_sacc = 0

for j in range(promote_len):
  #print(promote_seqs_strep[j] + "\n")
  for q in range(nuc_len-foxolen):
    if promote_seqs_strep[j][q: q + foxolen] == foxo1Consensus:
      motif_count_strep += 1
      #print("Match at S. coelicolor position", q,"in gene",j)
    if promote_seqs_sacc[j][q: q + foxolen] == foxo1Consensus:
      motif_count_sacc += 1
      #print("Match at S. cerevisiae position", q,"in gene",j)

print("S. coelicolor matches:",motif_count_strep,"\n")
print("S. cerevisiae matches:",motif_count_sacc,"\n")



```
